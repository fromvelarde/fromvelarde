```python
import numpy as np
import matplotlib.pyplot as plt

# Parmetros supuestos
A = 1.0e13  # s^-1
E = 8.0e4   # J/mol
R = 8.314   # J/(mol K)
c_in = 1.0  # Concentracin inicial
beta = 1.0  # Parmetro beta
alpha = 0.01  # Parmetro alpha
T_in = 300   # Temperatura inicial (K)

# Definimos la constante de velocidad k(T) usando la ley de Arrhenius
def k(T):
    return A * np.exp(-E / (R * T))

# Definimos las curvas
def y1(T):
    return c_in * beta * k(T) / (1 + alpha * k(T))

def y2(T):
    return T - T_in

# Creamos un rango de temperaturas
T_values = np.linspace(300, 500, 400)

# Calculamos los valores de y para cada curva
y1_values = y1(T_values)
y2_values = y2(T_values)

# Graficamos
plt.figure(figsize=(12, 6))
plt.plot(T_values, y1_values, label='y = c_in * beta * k(T) / (1 + alpha * k(T))')
plt.plot(T_values, y2_values, label='y = T - T_in')
plt.xlabel('Temperatura (K)')
plt.ylabel('y')
plt.title('Grficas de las curvas dadas')
plt.legend()
plt.grid(True)
plt.show()

#####################Variacin de signos

# Parmetros supuestos
A = 1.0e13  # s^-1
E = 8.0e4   # J/mol
R = 8.314   # J/(mol K)
c_in = 1.0  # Concentracin inicial
beta = 1.0  # Parmetro beta
alpha = 0.01  # Parmetro alpha
T_in = 400   # Temperatura inicial (K)

# Definimos la constante de velocidad k(T) usando la ley de Arrhenius
def k(T):
    return A * np.exp(-E / (R * T))

# Definimos las curvas
def y1(T):
    return c_in * beta * k(T) / (1 + alpha * k(T))

def y2(T):
    return T - T_in

# Creamos un rango de temperaturas
T_values = np.linspace(300, 500, 400)

# Calculamos los valores de y para cada curva
y1_values = y1(T_values)
y2_values = y2(T_values)

# Graficamos
plt.figure(figsize=(12, 6))
plt.plot(T_values, y1_values, label='y = c_in * beta * k(T) / (1 + alpha * k(T))')
plt.plot(T_values, y2_values, label='y = T - T_in')
plt.xlabel('Temperatura (K)')
plt.ylabel('y')
plt.title('Grficas de las curvas dadas')
plt.legend()
plt.grid(True)
plt.show()

# la otra curva

import numpy as np
import matplotlib.pyplot as plt

# Definimos la constante de los gases ideales R
R = 8.314  # J/(mol K)

# Asumimos valores para la constante de velocidad pre-exponencial k0 y la energa de activacin E
k0 = 7.0e10  # 1/s
E = 8.0e4  # J/mol
alpha = 0.01  # Asumimos un valor para alpha

# Definimos la funcin T' segn la ecuacin (6)
def T_prime(T):
    return (k0 * np.exp(-E / (R * T))) / (1 + alpha * k0 * np.exp(-E / (R * T)))

# Generamos un rango de temperaturas para graficar
T_values = np.linspace(300, 1000, 400)  # Kelvin
T_prime_values = T_prime(T_values)

# Graficamos la funcin
plt.figure(figsize=(10, 6))
plt.plot(T_values, T_prime_values, label="T' vs T")
plt.xlabel('Temperatura (K)')
plt.ylabel("T' (derivada de la temperatura respecto al tiempo)")
plt.title("Grfico de T' segn la ecuacin sigmoidea")
plt.legend()
plt.grid(True)
plt.show()
```
