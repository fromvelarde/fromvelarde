```python

def f_x(x):
    y = zeros(len(x))
    for i in range(len(x)):
        if -1 <= x[i] <= 0:
            y[i] = 1
        else:  # asumimos que cualquier valor no entre -1 y 0 será entre 0 y 1
            y[i] = 0
    return y
x = linspace(-1, 1, 2000)  # Ajustamos el rango de x para que vaya de -1 a 1
f = f_x(x)
```

```python
plot(x,f,label='$f(x)$')
plt.xlabel('$x$'); plt.ylabel('$f(x)$');
plt.title("Condición inicial (u(x,0)=f(x))")
xlim([-1.01, 1.01])
ylim([-0.01, 1.5]),grid(),legend()
show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import math

# Definir la primera función por partes
def f(x):
    if -2 <= x < 0:
        return -1
    elif 0 <= x < 2:
        return 2 - x
    else:
        return None

# Definir la segunda función
def funcion_completa(x):
    result = 0
    for n in range(1, 21):
        numerator_1 = 2 * (1 - (-1)**n) * math.cos((math.pi * n * x) / 2)
        denominator_1 = (math.pi**2) * (n**2)
        term1 = numerator_1 / denominator_1

        numerator_2 = (3 - (-1)**n) * math.sin((math.pi * n * x) / 2)
        denominator_2 = math.pi * n
        term2 = numerator_2 / denominator_2

        result += term1 + term2

    return result

# Crear un conjunto de puntos x
x_values = np.linspace(-2, 2, 1000)

# Aplicar la primera función a cada punto en x
y1_values = np.array([f(i) for i in x_values])

# Calcular los valores de la segunda función para cada punto x
y2_values = np.array([funcion_completa(x) for x in x_values])

# Graficar ambas funciones superpuestas
plt.figure(figsize=(8, 6))
plt.plot(x_values, y1_values, label='f(x) - Función por partes', color='blue')
plt.plot(x_values, y2_values, label='Función completa', linestyle='--', color='red')

# Configurar la gráfica
plt.title('Gráfica de ambas funciones')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()

# Mostrar la gráfica
plt.show()
```

La serie de fourier:

```python
import math
import numpy as np
import matplotlib.pyplot as plt

# Definir la función
def calcular_funcion(x):
    constante = (-1 + math.exp(4)) / (4 * math.exp(2))
    suma = 0
    for n in range(1, 21):
        term1 = (4 * ((-1)**n) * math.cos((math.pi * n * x) / 2) * math.sinh(2)) / ((math.pi**2 * n**2) + 4)
        term2 = (-2 * ((-1)**n) * math.pi * n * (math.sin((math.pi * n * x) / 2)) * math.sinh(2)) / ((math.pi**2 * n**2) + 4)
        suma += term1 + term2
    return constante + suma

# Crear un conjunto de puntos x
x_values = np.linspace(-2, 2, 1000)

# Calcular los valores de y para cada punto x
y_values = [calcular_funcion(x) for x in x_values]

# Graficar la función
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label='f(x)')

# Configurar la gráfica
plt.title('Gráfica de la serie de fourier')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()

# Mostrar la gráfica
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import math

# Definir la función f(x) = e^x
def f(x):
    return np.exp(x)

# Crear un conjunto de puntos x
x_values = np.linspace(-2, 2, 1000)

# Calcular los valores de y para la primera función
y_values_1 = f(x_values)

# Definir la función compuesta
def calcular_funcion(x):
    constante = (-1 + math.exp(4)) / (4 * math.exp(2))
    suma = 0
    for n in range(1, 21):
        term1 = (4 * ((-1)**n) * math.cos((math.pi * n * x) / 2) * math.sinh(2)) / ((math.pi**2 * n**2) + 4)
        term2 = (-2 * ((-1)**n) * math.pi * n * (math.sin((math.pi * n * x) / 2)) * math.sinh(2)) / ((math.pi**2 * n**2) + 4)
        suma += term1 + term2
    return constante + suma

# Calcular los valores de y para la segunda función
y_values_2 = [calcular_funcion(x) for x in x_values]

# Graficar ambas funciones en la misma figura
plt.figure(figsize=(8, 6))

# Graficar la primera función
plt.plot(x_values, y_values_1, label='f(x) = e^x')

# Graficar la segunda función
plt.plot(x_values, y_values_2, label='f(x) compuesta', linestyle='--')

# Configurar la gráfica
plt.title('Grafica de ambas funciones')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()

# Mostrar la gráfica combinada
plt.show()
```
